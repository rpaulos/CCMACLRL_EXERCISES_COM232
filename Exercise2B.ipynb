## Building Different Regression Models

### Model 1: Miles Traveled and Travel Time

16. Set independent and dependent variables

```python
# put your answer here
X = df[['Miles Traveled']]
X  = sm.add_constant(X)
y = df['Travel Time']
```

17. Build and train the model

```python
# put your answer here
mlr_model = LinearRegression()

mlr_model.fit(X, y)

y_pred = mlr_model.predict(X)
print('Intercept: ', mlr_model.intercept_)
print('Coefficients: ', mlr_model.coef_)
```

18. Display the statistical summary

```python
# put your answer here
olsmod = sm.OLS(y, X).fit()
olsmod.summary()
```

19. Display the r-squared

```python
print(olsmod.rsquared)
```

### Most Optimal Model (Mils Traveled, Gas Price, Number of Deliveries, and Travel Time)

```python
# put your answer here
X = df[['Miles Traveled', 'Gas Price', 'Number of Deliveries']]
X  = sm.add_constant(X)
y = df['Travel Time']
```
